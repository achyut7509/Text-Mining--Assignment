# 6. EMOTION MINING :

### 6.1 Data Preparation :

```python
cleaned_review_data_copy = cleaned_review_data.copy()
```

```python
cleaned_review_data_copy.columns
```

```python
cleaned_review_data_copy.drop(labels=['id', 'name', 'asins', 'brand', 'categories', 'manufacturer','reviews.rating'],axis=1,inplace=True)
cleaned_review_data_copy.head(10)
```

```python
cleaned_review_data_copy.columns=['reviews_text']
cleaned_review_data_copy.head()
```

### 6.2 Data Pre-processing:

```python
 # remove both the leading and the trailing characters
cleaned_review_data_copy=[reviews_text.strip() for reviews_text in cleaned_review_data_copy.reviews_text]
cleaned_review_data_copy[0:]
```

```python
 # removes empty strings, because they are considered in Python as False
cleaned_review_data_copy=[reviews_text for reviews_text in cleaned_review_data_copy if reviews_text]
cleaned_review_data_copy[0:]
```

```python
from nltk import tokenize
sentences =tokenize.sent_tokenize(" ".join(cleaned_review_data_copy))
sentences[0:]
```

```python
sentences_df = pd.DataFrame(sentences, columns=['Review_Texts'])
sentences_df
```

```python
afinn = pd.read_csv('Afinn.csv', sep=',',encoding='Latin-1')
afinn
```

```python
afinn.shape
```

```python
affinity_scores = afinn.set_index('word')['value'].to_dict()
affinity_scores
```

```python
#Custom function :score each word in a sentence in lemmatised form, 
#but calculate the score for the whole original sentence.
nlp = spacy.load('en_core_web_sm')
sentiment_lexicon = affinity_scores

def calculate_sentiment(review_text: str = None):
    sent_score = 0
    if review_text:
        sentence = nlp(review_text)
        for word in sentence:
            sent_score += sentiment_lexicon.get(word.lemma_, 0)
    return sent_score
```

```python
# test that it works
calculate_sentiment(review_text = 'amazing')
```

```python
calculate_sentiment(review_text = 'disappointing')
```

```python
sentences_df['sentiment_value'] = sentences_df['Review_Texts'].apply(calculate_sentiment)
sentences_df
```

```python
sentences_df.head(30)
```

```python
# how many words are in the Review_Text?
sentences_df['word_count'] = sentences_df['Review_Texts'].str.split().apply(len)
sentences_df['word_count'].head(50)
```

```python
sentences_df.sort_values(by='sentiment_value').head(25)
```
